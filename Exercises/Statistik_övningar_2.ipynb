**A sensor in a smart network has a probability $p = .7$ to correctly detect a person entering a room.**

**a) If 10 people enter the room, how many detections are expected?**

**b) If 12 people enter the room, what is the probability that between 2 and 7 people were deteted?**

**c) If 3 people were detected, what is the likelyhood 5 people having entered the room?**

a) 7 då sannolikheten är $\frac{7}{10}$

b) Då det är oberoende händelser kan vi använda oss av binomialfördelning $P[X=k]=\binom{n}{k}p^k(1-p)^{n-k}$ för spannet 2-7 och summera resultaten:

$P[2 \le X \le 7] = P[X=2] + P[X=3] + P[X=4] + P[X=5] + P[X=6] + P[X=7]$

$P[X=2]=\binom{12}{2}(\frac{7}{10})^2(1-\frac{7}{10})^{12-2} =\ _{12}C_2(\frac{7}{10})^2(1-\frac{7}{10})^{12-2} = \frac{\frac{12!}{(12-2)!}}{2!}(\frac{7}{10})^2(1-\frac{7}{10})^{12-2} = 66 \cdot .49 \cdot .0000059049 \approx .000191$

$P[X=3]=\binom{12}{3}(\frac{7}{10})^3(1-\frac{7}{10})^{12-3} =\ _{12}C_3(\frac{7}{10})^3(1-\frac{7}{10})^{12-3} = \frac{\frac{12!}{(12-3)!}}{3!}(\frac{7}{10})^3(1-\frac{7}{10})^{12-3} = 220 \cdot .343 \cdot .000019683 \approx .001485$

$P[X=4]=\binom{12}{4}(\frac{7}{10})^4(1-\frac{7}{10})^{12-4} =\ _{12}C_4(\frac{7}{10})^4(1-\frac{7}{10})^{12-4} = \frac{\frac{12!}{(12-4)!}}{4!}(\frac{7}{10})^4(1-\frac{7}{10})^{12-4} = 495 \cdot .2401 \cdot .00006561 \approx .007798$

$P[X=5]=\binom{12}{5}(\frac{7}{10})^5(1-\frac{7}{10})^{12-5} =\ _{12}C_5(\frac{7}{10})^5(1-\frac{7}{10})^{12-5} = \frac{\frac{12!}{(12-5)!}}{5!}(\frac{7}{10})^5(1-\frac{7}{10})^{12-5} = 792 \cdot .16807 \cdot .0002187 \approx .029112$

$P[X=6]=\binom{12}{6}(\frac{7}{10})^6(1-\frac{7}{10})^{12-6} =\ _{12}C_6(\frac{7}{10})^6(1-\frac{7}{10})^{12-6} = \frac{\frac{12!}{(12-6)!}}{6!}(\frac{7}{10})^6(1-\frac{7}{10})^{12-6} = 924 \cdot .117649 \cdot .000729 \approx .079248$

$P[X=7]=\binom{12}{7}(\frac{7}{10})^7(1-\frac{7}{10})^{12-7} =\ _{12}C_7(\frac{7}{10})^7(1-\frac{7}{10})^{12-7} = \frac{\frac{12!}{(12-7)!}}{7!}(\frac{7}{10})^7(1-\frac{7}{10})^{12-7} = 792 \cdot .0823543 \cdot .00243 \approx .158496$

$P[2 \le X \le 7] = .000191 + .001485 + .007798 + .029112 + .079248 + .158496 = .27633$

c) För att få reda på sannolikheten att 3 av fem uppfattades kan vi använda oss av sannolikhetsformeln för binomialfördelning $L[p|n,x] = \binom{n}{x}p^x(1-p)^{n-x}$ med n=5, x=3 och p=.7

$L[\frac{7}{10}|5, 3] = \binom{5}{3}(\frac{7}{10})^3(1-\frac{7}{10})^{5-3} = 10 \cdot .343 \cdot .09 = .3087$

$L[\frac{7}{10}|5, 3] \approx 31\%$
