**A sensor in a smart network has a probability $p = .7$ to correctly detect a person entering a room.**

**a) If 10 people enter the room, how many detections are expected?**

**b) If 12 people enter the room, what is the probability that between 2 and 7 people were deteted?**

**c) If 3 people were detected, what is the likelyhood 5 people having entered the room?**

a) 7 då sannolikheten är $E[X] = \mu = np = 10 \cdot .7 = 7$

b)  
Då det är oberoende händelser kan vi använda oss av binomialfördelning  
$P[X=k]=\binom{n}{k}p^k(1-p)^{n-k}$  
för spannet 2-7 och summera resultaten:  
$P[2 \le X \le 7] = P[X=2] + P[X=3] + P[X=4] + P[X=5] + P[X=6] + P[X=7]$  
$P[X=2]=\binom{12}{2}(\frac{7}{10})^2(1-\frac{7}{10})^{12-2} =\ _{12}C_2(\frac{7}{10})^2(1-\frac{7}{10})^{12-2} = \frac{\frac{12!}{(12-2)!}}{2!}(\frac{7}{10})^2(1-\frac{7}{10})^{12-2} = 66 \cdot .49 \cdot .0000059049 \approx .000191$  
$P[X=3]=\binom{12}{3}(\frac{7}{10})^3(1-\frac{7}{10})^{12-3} =\ _{12}C_3(\frac{7}{10})^3(1-\frac{7}{10})^{12-3} = \frac{\frac{12!}{(12-3)!}}{3!}(\frac{7}{10})^3(1-\frac{7}{10})^{12-3} = 220 \cdot .343 \cdot .000019683 \approx .001485$  
$P[X=4]=\binom{12}{4}(\frac{7}{10})^4(1-\frac{7}{10})^{12-4} =\ _{12}C_4(\frac{7}{10})^4(1-\frac{7}{10})^{12-4} = \frac{\frac{12!}{(12-4)!}}{4!}(\frac{7}{10})^4(1-\frac{7}{10})^{12-4} = 495 \cdot .2401 \cdot .00006561 \approx .007798$  
$P[X=5]=\binom{12}{5}(\frac{7}{10})^5(1-\frac{7}{10})^{12-5} =\ _{12}C_5(\frac{7}{10})^5(1-\frac{7}{10})^{12-5} = \frac{\frac{12!}{(12-5)!}}{5!}(\frac{7}{10})^5(1-\frac{7}{10})^{12-5} = 792 \cdot .16807 \cdot .0002187 \approx .029112$  
$P[X=6]=\binom{12}{6}(\frac{7}{10})^6(1-\frac{7}{10})^{12-6} =\ _{12}C_6(\frac{7}{10})^6(1-\frac{7}{10})^{12-6} = \frac{\frac{12!}{(12-6)!}}{6!}(\frac{7}{10})^6(1-\frac{7}{10})^{12-6} = 924 \cdot .117649 \cdot .000729 \approx .079248$  
$P[X=7]=\binom{12}{7}(\frac{7}{10})^7(1-\frac{7}{10})^{12-7} =\ _{12}C_7(\frac{7}{10})^7(1-\frac{7}{10})^{12-7} = \frac{\frac{12!}{(12-7)!}}{7!}(\frac{7}{10})^7(1-\frac{7}{10})^{12-7} = 792 \cdot .0823543 \cdot .00243 \approx .158496$  
$P[2 \le X \le 7] = .000191 + .001485 + .007798 + .029112 + .079248 + .158496 = .27633$

c)  
För att få reda på sannolikheten att tre av fem uppfattades kan vi använda oss av sannolikhetsformeln för binomialfördelning  
$L[p|n,x] = \binom{n}{x}p^x(1-p)^{n-x}$  
med n=5, x=3 och p=.7  
$L[\frac{7}{10}|5, 3] = \binom{5}{3}(\frac{7}{10})^3(1-\frac{7}{10})^{5-3} = 10 \cdot .343 \cdot .09 = .3087$  
$L[\frac{7}{10}|5, 3] \approx 31\%$  
d.v.s. att det är drygt 31% chans att tre av fem uppfattades

**A large random sample from an unknown distribution is drawn. the mean is calculated to be 0 and the standard deviation is 1.**

**a) What kind of distributions could the sample have been drawn from?**

**b) What is the probability that a normally distributed random variable will lie within one, two and three standard deviations from its mean?**

**Consider the smart network sensors from the previous exercise. Across 120 common rooms, a total of 24000 people move through an office complex per day.**

**c) How many detections are expected per day?**

**d) How many detections would be considered unusually few?**

a)  
Endast normaldistribution då medelvärdet  
$\mu = 0$  
och standardavvikelsen  
$\sigma = 1$  
blir  
$\frac{X-\mu}{\sigma} = \frac{X-0}{1} = X$  
och uppfyller därför  
$X \sim N(0,1) = Z$  
eftersom att  
$Z = \frac{X-\mu}{\sigma}$

b)  
$P(\mu - \sigma \le X \le \mu + \sigma) = P(0 - 1 \le \frac{X-\mu}{\sigma} \le 0 + 1) = P(-1 \le Z \le 1)$  
$P(\mu - 2\sigma \le X \le \mu + 2\sigma) = P(0 - 1 \cdot 2 \le \frac{X-\mu}{\sigma} \le 0 + 1 \cdot 2) = P(-2 \le Z \le 2)$   
$P(\mu - 3\sigma \le X \le \mu + 3\sigma) = P(0 - 1 \cdot 3 \le \frac{X-\mu}{\sigma} \le 0 + 1 \cdot 3) = P(-3 \le Z \le 3)$

c)  
$E[X] = \mu = np = 24000 \cdot .7 = 16800$

d)  
Medelvärdet  
$\mu = 16800$  
och standardavvikelsen  
$\sigma = \sqrt{np(1-p)} = \sqrt{16800 \cdot (1-.7)} \approx 71$  
ger oss att färre än $16658$ detections hade räknats som ovanligt få då det brukar anses vara $X < \mu - 2\sigma$

**During a heavy traffic scenario, it was found that the probability of successfully logging on to a network router at any point was 70%.**

**a) Is it unusal if two attempts are required to log in during heavy traffic?**

**b) What is the likelyhood of observing (exactly) three attempts?**

**c) At what point does the number of attempts become unusual?**

a)  
Med formeln för geometrisk fördelning  
$P[X=k] = (1-p)^{k-1} \cdot p$  
får vi  
$P[X=2] = (1-.7)^{2-1} \cdot .7 = (.3)^{1} \cdot .7 = .21$  
vilket innebär att det är 21% chans att det krävs två försök för att lyckas logga in, därav räknas det inte som ovanligt

Man hade även kunnat använda normalapproximation, för om  
$X \sim Geom(p)$  
gäller då  
$\mu=E[X]=\frac{1}{p}$  
och  
$\sigma=\sqrt{\frac{1-p}{p^2}}$  
vilket ger  
$\mu = \frac{1}{.7} \approx 1.42, \sigma \approx .78$  
och därmed  
$P(.65 \le X \le 2.21)$  
d.v.s. att alla värden mellan 0,65 och 2,21 anses vara inom spannet för det normala och alltså inte ovanligt

b)  
$P[X=3] = (1-.7)^{3-1} \cdot .7 = (.3)^{2} \cdot .7 = .063$  
ger oss att det är 6,3% chans att lyckas först på tredje försöket vilket innebär att man observerat exakt tre försök.

c)  
$P[X=4] = (1-.7)^{4-1} \cdot .7 = (.3)^{3} \cdot .7 = .0189$  
ger oss att det är 1,89% chans att lyckas först på fjärde försöket och eftersom att under 5% anses ovanligt så är det vid fyra eller fler försök som det anses vara ovanligt många försök

**As a follow up to the heavy traffic scenario, a system was devised that would automatically log in to a number of routers in order to retain monitoring and control in case of network congestion. 12 routers are part of the system. The logins are independent of eachother. The probability of a successful login attempt remains at £p = .7$.**

**a) How many attempts are expected before all 12 systems are connected?**

**b) What is the probability of needing exactly 12 attempts?**

**c) In a reasonable worst-case scenario, how many attempts could be expected to be required?**

a)  
Med negativ binomialfördelning  
$E[X]=\frac{r}{p}$  
får vi  
$E[X] = \frac{12}{.7} \approx 17.14$  
d.v.s. att det förväntas krävas precis över 17 försök, så alltså 18 försök eftersom att man inte kan delvis försöka logga in

b)  
För att exakt 12 försök behövs innebär det alla tolv routrars försök att logga in lyckas på första försöket, d.v.s.  
$P[X=12] = (.7)^{12} \approx .0138$  
vilket innebär att det är ungefär 1,38% chans att alla tolv routrars försök att logga in lyckas på första försöket

c)  
Vad som anses resonligt är väl subjektivt men om vi räknar med att det vore två standardavvikelser från medelvärdet så kan vi nå det svaret med följande uträkning  
$\sigma = \sqrt{\frac{r(1-p)}{p^2}} = \sqrt{\frac{12 \cdot .3}{.49}} \approx 2.71$  
och då med  
$E[X] = \mu \approx 17.14$  
som vi fick från a) gäller då  
$\mu + 2\sigma \approx 17.14 + 2 \cdot 2.71 \approx 22.56$
d.v.s. att 23 eller fler försök hade ansets vara ett resonligt värsta scenario
