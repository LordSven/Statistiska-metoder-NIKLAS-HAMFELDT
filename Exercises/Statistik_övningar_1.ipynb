## 2

**When a woman is a carrier of classical hemophelia, there is a 50% chance that a male child will inherit the disease. If a carrier gives birth to two sons, what is the probability that both boys will have the disease? What approach are you using to answer the question?**

$P[A_1 \cap A_2] = P(A_1) \cdot P(A_2)$

$A_1 = 0.5$

$A_2 = 0.5$

$0.5 \cdot 0.5 = 0.25$

$P[A_1 \cap A_2] = 25\%$

$A_1$ är första sonen och $A_2$ är andra sonen, då de är oberoende av varandra och båda har 50% sannolikhet att få sjukdomen så får vi genom att avnända snittet av de två händelserna att det är 25% sannolikhet att båda får den.

Vill man räkna ut hur stor sannolikheten är att någon av sönerna får sjukdomen så kan man göra det genom följande metod (unionen):

$P[A_1 \cup A_2] = P(A_1) + P(A_2) - P[A_1 \cap A_2]$

$0.5 + 0.5 - 0.25 = 0.75$

$P[A_1 \cup A_2] = 75\%$
